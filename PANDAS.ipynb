# JSON

```python
import json
# json object only accept string, byte or bytearray
json_data= """{"name":"dev","phone":"1234","age":"21","address":["shatabdi nagar","panchvati enclave"]}"""

type(json_data)
```

```python
loading=json.loads(json_data)    ##converts string, byte data into json format

loading
```

```python
pd.DataFrame(loading)
```
